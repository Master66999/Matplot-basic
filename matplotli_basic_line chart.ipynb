```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3] ,[4,5,6])
plt.show()
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years, runs)
plt.show()
```

```python
plt.ylabel("Runs")
plt.xlabel("year")
plt.title("sachine score")
plt.plot(years,runs)
plt.show()
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years,kohli , label = "kohil runs")
plt.plot(years,sehwag , label = "sehwag runs")
plt.legend()
plt.show()
```

```python
plt.plot(years, kohli, 'g^:', label="Kohli")  # red circles with dashed lines
plt.plot(years, sehwag, 'ro--', label="Sehwag")  # green triangles dotted
plt.legend()
```

```python
plt.plot(years, kohli , 'ro--',label = "kohli")
plt.plot(years , sehwag , 'g^:', label = "sehwag" )
plt.legend()
```

```python
plt.plot(years, kohli, color='orange', linestyle='--', label="Kohli")
plt.plot(years, sehwag, color='green', linestyle='-.', label="Sehwag")
plt.plot(years, runs, color='blue', label="Tendulkar")
plt.legend()
```

```python
plt.plot(years, kohli, linewidth=3, label="Kohli")
plt.plot(years, sehwag, linewidth=2, label="Sehwag")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
with plt.xkcd():
    plt.plot(years, kohli, label="Kohli")
    plt.plot(years, sehwag, label="Sehwag")
    plt.title("Epic Battle of the Batsmen")
    plt.legend()
    plt.show()
```
