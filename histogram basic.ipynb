```python
import matplotlib.pyplot as plt
import numpy as np

ages = [
    34, 28, 36, 45, 27, 27, 45, 37, 25, 35,
    25, 25, 32, 10, 12, 24, 19, 33, 20, 15,
    44, 27, 30, 15, 24, 31, 18, 33, 23, 27,
    23, 48, 29, 19, 38, 17, 32, 10, 16, 31,
    37, 31, 28, 26, 15, 22, 25, 40, 33, 12,
    33, 26, 23, 36, 40, 39, 21, 26, 33, 39,
    25, 28, 18, 18, 38, 43, 29, 40, 33, 23,
    33, 45, 29, 45, 3, 38, 30, 27, 30, 10,
    27, 33, 44, 24, 21, 24, 39, 33, 24, 35,
    30, 39, 22, 26, 26, 15, 32, 32, 30, 27
]

bins = [10, 20, 30, 40, 50, 60, 70]

plt.hist(ages, bins=bins, edgecolor='black')
plt.title('Age Distribution of YouTube Viewers')
plt.xlabel('Age Group')
plt.ylabel('Number of Viewers')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.hist(ages, bins=bins, edgecolor='black')
plt.axvline(np.mean(ages), color='red', linestyle='--', linewidth=2, label='Average Age')
plt.legend()
plt.title('Age Distribution with Mean Line')
plt.show()
```

### Example 1: Histogram with different color and cumulative distribution

```python
plt.hist(ages, bins=bins, color='skyblue', edgecolor='black', cumulative=True)
plt.title('Cumulative Age Distribution of YouTube Viewers')
plt.xlabel('Age Group')
plt.ylabel('Cumulative Number of Viewers')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### Example 2: Histogram with logarithmic scale and custom binning

```python
log_bins = np.arange(0, 50, 5) # Create bins from 0 to 50 with a step of 5

plt.hist(ages, bins=log_bins, color='lightgreen', edgecolor='black', log=True)
plt.title('Age Distribution of YouTube Viewers (Log Scale)')
plt.xlabel('Age Group')
plt.ylabel('Number of Viewers (Log Scale)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### Example 3: Stacked Histogram with two age groups

```python
mean_age = np.mean(ages)

younger_viewers = [age for age in ages if age < mean_age]
older_viewers = [age for age in ages if age >= mean_age]

plt.hist([younger_viewers, older_viewers], bins=bins, stacked=True, edgecolor='black', color=['lightcoral', 'lightskyblue'], label=['Younger Viewers', 'Older Viewers'])
plt.title('Stacked Age Distribution of YouTube Viewers')
plt.xlabel('Age Group')
plt.ylabel('Number of Viewers')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### Example 4: Histogram with Normalized Counts (Density)

```python
plt.hist(ages, bins=bins, edgecolor='black', color='lightsalmon', density=True)
plt.title('Age Distribution of YouTube Viewers (Density)')
plt.xlabel('Age Group')
plt.ylabel('Probability Density')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### Example 5: Histogram with different histtype and custom number of bins

```python
plt.hist(ages, bins=15, histtype='stepfilled', edgecolor='black', color='orchid', alpha=0.7)
plt.title('Age Distribution of YouTube Viewers (Stepfilled)')
plt.xlabel('Age Group')
plt.ylabel('Number of Viewers')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### Example 6: Histogram with custom colors for each bar

```python
n, bins, patches = plt.hist(ages, bins=bins, edgecolor='black')

# Color bars based on their height (or any other criteria)
for i, patch in enumerate(patches):
    if i % 2 == 0:
        patch.set_facecolor('teal')
    else:
        patch.set_facecolor('steelblue')

plt.title('Age Distribution of YouTube Viewers (Custom Bar Colors)')
plt.xlabel('Age Group')
plt.ylabel('Number of Viewers')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### Example 7: Overlaid Histograms (comparing younger vs. older viewers)

```python
plt.hist(younger_viewers, bins=bins, edgecolor='black', color='red', alpha=0.5, label='Younger Viewers')
plt.hist(older_viewers, bins=bins, edgecolor='black', color='blue', alpha=0.5, label='Older Viewers')
plt.title('Overlaid Age Distribution (Younger vs. Older)')
plt.xlabel('Age Group')
plt.ylabel('Number of Viewers')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
